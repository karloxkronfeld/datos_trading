```python
import pandas as pd
from pylab import *
```

```python
datos=pd.read_csv("SEPTIEMBRE_XAUUSD.xlsx")
datos=datos[["time","bid","ask"]]
```

```python
datos["precio"]=(datos.bid+datos.ask)/2
datos=datos.groupby("time").mean()
datos.index=pd.DatetimeIndex(datos.index)
datos.index=datos.index.strftime("%Y-%m-%d %H:%M")
minuto=datos.groupby("time").mean()
```

```python
minuto["MA200"]=minuto.precio.rolling(200).mean()
```

```python
minuto["diferencia"]=minuto.precio-minuto.MA200
```

```python
minuto[["precio","MA200"]].plot(figsize=(20,7))
twinx()
minuto.diferencia.plot(color="red")
```
